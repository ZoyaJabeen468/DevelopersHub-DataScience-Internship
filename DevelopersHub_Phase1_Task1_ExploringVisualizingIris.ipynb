## Task 1: Exploring and Visualizing a Simple Dataset

**Objective:** Understand how to read, summarize, and visualize a dataset
using pandas, matplotlib, and seaborn.

**Dataset:** Iris Dataset
- 150 rows, 5 columns
- 4 numeric features (sepal length, sepal width, petal length, petal width)
- Target: species (3 balanced classes — setosa, versicolor, virginica)

**Approach:**
1. Load the dataset and inspect its structure (`.shape`, `.columns`, `.head()`)
2. Check for missing values and class balance
3. Visualize relationships with a scatter plot
4. Examine distributions with histograms
5. Detect spread and outliers with box plots
6. Summarize feature relationships with a pair plot

```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')
```

```python
# Loading the Iris dataset directly via seaborn
iris = sns.load_dataset('iris')
```

```python
# Shape: number of rows and columns
print("Shape:", iris.shape)

# Column names
print("\nColumns:", iris.columns.tolist())

# First 5 rows
iris.head()
```

```python
# Summary statistics for all numeric columns
iris.describe()
```

```python
# Check for missing values
print("Missing values:\n", iris.isnull().sum())

# Class distribution
print("\nSpecies count:\n", iris['species'].value_counts())
```

## Visualizations

```python
# Scatter plot: petal length vs petal width, colored by species
plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', palette='Set2', s=80)
plt.title('Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
# Histogram of all four numeric features, one per species
iris.hist(figsize=(10, 7), bins=20, color='steelblue', edgecolor='white')
plt.suptitle('Distribution of Iris Features', y=1.02, fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# Overlapping histograms per species for sepal length
plt.figure(figsize=(8, 5))
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    plt.hist(subset['sepal_length'], alpha=0.6, label=species, bins=15)
plt.title('Sepal Length Distribution by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Box plots for all four features split by species
fig, axes = plt.subplots(2, 2, figsize=(11, 8))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for ax, feature in zip(axes.flatten(), features):
    sns.boxplot(data=iris, x='species', y=feature, palette='Set2', ax=ax)
    ax.set_title(f'{feature.replace("_", " ").title()} by Species')
    ax.set_xlabel('Species')

plt.suptitle('Box Plots - Iris Features by Species', fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Pair plot showing all feature relationships at once
sns.pairplot(iris, hue='species', palette='Set2', diag_kind='kde')
plt.suptitle('Pair Plot — Iris Dataset', y=1.02)
plt.show()
```

## Conclusion

- The Iris dataset has 150 samples, 4 numeric features, and 3 balanced classes (50 each).
- No missing values were found - the dataset is clean and ready to use.
- Scatter plots show that petal_length and petal_width together almost perfectly separate the three species.
- Histograms confirm that Setosa has a distinctly different (narrower) petal distribution.
- Box plots reveal that Virginica tends to have the largest measurements, and Setosa the smallest petals.
- The dataset is well-suited for classification tasks due to its clear inter-species separability in petal features.
